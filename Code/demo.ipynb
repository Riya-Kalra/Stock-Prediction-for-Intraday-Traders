```python
import yfinance as yf
from datetime import datetime
import pandas as pd

data = yf.download(tickers='AAPL', 
                    period='2d', 
                    interval='15m',
                    auto_adjust = True)
```

```python
# add time column
time_column = []
for i in data.index:
    time_column.append(i.to_pydatetime())

data = data.assign(time=pd.Series(time_column).values)
```

```python
data.reset_index(inplace=True)

# rename columns according to our older convention
data.rename(
    { 'Open': 'open',
    'High': 'high',
    'Low' : 'low',
    'Close':'close',
    'Volume' : 'volume'},
    axis=1,
    inplace=True)
```

```python
# drop rows where volume traded==0 it only happens during start of day 
# so is probably not needed
drop_rows = []
for i in range(len(data)):
    # print(type(data["volume"][i]))
    if data["volume"][i]==0 :
        drop_rows.append(i)

data.drop(data.index[drop_rows], inplace=True)
```

```python
data.drop(['Datetime'], axis=1, inplace=True)
data
```

```python
history = df.to_dict(orient='records')
history
```

```python
time_format = "%Y-%m-%d %H:%M:%S"
today = []
today_index = None
for i in range(len(history)):
  if datetime.strptime(history[i]["time"], time_format).day == 25:
    today_index = i
    break

today = history[today_index:]
history = history[:today_index]
```

```python
# EMA
ema_points = 10
ema_multiplier = 2/(ema_points+1)
ema_yesterday = None
# VWAP
sum_prices = 0
previous_day = -1
sum_pv = 0
sum_volume = 0
# RSI
rsa_lookback = 14
prev_avg_gain = 0
prev_avg_loss = 0
prev_price = None
```

```python

################ startup RSI first ##############################################
lookback = [x["close"] for x in history[-rsa_lookback:]]
print(len(lookback))
previous = lookback[0]
gain = 0
loss = 0
for i in range(1,rsa_lookback):
    current = lookback[i]
    if current-previous > 0:
        gain += current-previous
    else:
        loss += previous-current 
    previous = current
prev_avg_gain = gain/rsa_lookback
prev_avg_loss = loss/rsa_lookback
prev_price = lookback[-1]
################### Startup EMA ##################################################
ema_yesterday = [x["close"] for x in history[-ema_points:]]
ema_yesterday = sum(ema_yesterday)/ema_points

print("Historical data loaded up .........................")
```

```python
##################################### EMA Calculation ######################################
for row in today:
    ema_today =  row["close"]*ema_multiplier + ema_yesterday*(1-ema_multiplier)  
    row["ema"] = ema_today
    ema_yesterday = ema_today

############################## VWAP calculation ##############################################

for row in today:
    current_day = datetime.strptime(row["time"], time_format).day

    if(current_day!=previous_day):
        sum_pv = 0
        sum_volume = 0        

    previous_day = current_day

    sum_pv += ((row["high"] + row["low"] + row["close"])/3) * row["volume"]
    sum_volume += row["volume"]
    row["vwap"] = sum_pv/sum_volume

############################## Add RSI ######################################################

for row in today:
    
    current_diff = row["close"]-prev_price
    prev_price = row["close"]

    if  current_diff > 0:
        avg_gain = (prev_avg_gain*13 + current_diff)/14
        avg_loss = (prev_avg_loss*13)/14
    else:
        avg_gain = (prev_avg_gain*13)/14
        avg_loss = (prev_avg_loss*13 - current_diff)/14
    
    prev_avg_gain = avg_gain
    prev_avg_loss = avg_loss

    rs = avg_gain/avg_loss
    row["rsi"] = 100 - 100/(1+rs)
```
