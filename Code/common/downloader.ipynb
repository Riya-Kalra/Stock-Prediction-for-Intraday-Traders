```python
import pandas as pd
import time

base_url = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY_EXTENDED&symbol={stock_symbol}&interval=15min&slice={duration}&apikey=8PHWP0VWMDDH4PJ3'
symbols = ["AAPL","MSFT","AMZN","TSLA","GOOGL","NVDA","BRK-B","FB"]

for stock_symbol in symbols:
  print(f"stock symbol {stock_symbol}")
  df_list = []
  for year in range(1,0,-1):
    for month in range(12,0,-1):
      duration = f"year{year}month{month}"
      print(duration)
      url = base_url.format(stock_symbol=stock_symbol, duration = duration)
      df = pd.read_csv(url)
      # sleep is required because the number of API calls per minute is restricted
      time.sleep(15)
      print(len(df))
      df_list.append(df)

  df_list_reverse = [x[::-1] for x in df_list]
  result_reverse = pd.concat(df_list_reverse,ignore_index=True)
  # change the path accordingly
  result_reverse.to_csv(f"/content/drive/MyDrive/Major Project 8th sem/Code/data/{stock_symbol}_1years_15Min.csv")
  print(len(result_reverse))
```
