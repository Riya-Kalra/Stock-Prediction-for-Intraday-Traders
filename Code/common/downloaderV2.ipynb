```python
stock_symbol = "RELIANCE.NS"
period = "60d"
interval = "15m"
```

```python
import yfinance as yf
from datetime import datetime
import pandas as pd

data = yf.download(tickers=stock_symbol, 
                    period=period, 
                    interval=interval,
                    auto_adjust = True)
```

```python
# add time column
time_column = []
for i in data.index:
    time_column.append(i.to_pydatetime())

data = data.assign(time=pd.Series(time_column).values)
```

```python
data.reset_index(inplace=True)

# rename columns according to our older convention
data.rename(
    { 'Open': 'open',
    'High': 'high',
    'Low' : 'low',
    'Close':'close',
    'Volume' : 'volume'},
    axis=1,
    inplace=True)
```

```python
# drop rows where volume traded==0 it only happens during start of day 
# so is probably not needed
drop_rows = []
for i in range(len(data)):
    # print(type(data["volume"][i]))
    if data["volume"][i]==0 :
        drop_rows.append(i)

data.drop(data.index[drop_rows], inplace=True)
```

```python
data.drop(['Datetime'], axis=1, inplace=True)
data
```

```python
data.to_csv(f"./{stock_symbol}_{period}_{interval}.csv")
data.isnull().sum()
```
